# Matplotlib

````python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
````

## Plot Types
```python
- plot()
    plot(kind='line')
    plot(kind='bar')
    plot(kind='barh')
    plot(kind='pie')
```

````python
olympics = pd.read_csv('olympics.csv', skiprows = 4)
````

__What were the different sports in the first olympics? Plot them using different graphs.__

````python
first = olympics[olympics.Edition == 1896]
````

__Line plot__

````python
first.Sport.value_counts()
````

````python
first.Sport.value_counts().plot();
````

__Bar plot__

````python
first.Sport.value_counts().plot(kind='bar');
````

__Horizontal bar plot__

````python
first.Sport.value_counts().plot(kind='barh');
````

__Pie Chart__

````python
first.Sport.value_counts().plot(kind='pie', figsize=(6,6));
````

## Plot Colors
```python
plot(kind='bar', color='blue')
```

````python
first.Sport.value_counts().plot(kind='bar', color='red');
````

## Figure Size
```python
plot(figsize=(width,height))
    plot(kind='bar', color='yellow', figsize(5,5))
```

````python
first.Sport.value_counts().plot(figsize=(9,4));
````

## Colormaps
__Classes of colormaps__
- Sequential
- Diverging
- Qualitative

````python
first.Sport.value_counts().plot(kind='pie', colormap='Pastel2', figsize=(5,5));
````

# Seaborn Basic Ploting
- __countplot()__

````python
import seaborn as sns
````

__How many medals have been won by men and women in the history of the Olympics? How many gold, silver, and bronze medals were won for each gender?__

````python
sns.countplot(x='Medal', data=olympics);
````

````python
sns.countplot(x='Medal', data=olympics, hue='Gender');
````

## Challenge

__Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using both `Matplotlib` and `Seaborn`.__

````python
cn = olympics[(olympics.Edition == 2008) & (olympics.NOC == 'CHN')]
````

````python
cn.Gender.value_counts()
````

````python
cn.Gender.value_counts().plot(kind='bar');
````

````python
plt.figure(figsize=(7,7))
sns.countplot(x='Gender', data=cn);
````

__Plot the number of gold, silver, and bronze medals for each gender__

````python
sns.countplot(x='Medal', hue='Gender', data = cn);
````

````python
sns.countplot(x='Medal', hue='Gender', data=cn, order=['Gold','Silver','Bronze']);
````
