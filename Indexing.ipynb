# Index
- The index object is an immutable array
- Indexing allows you to access a row or column using a label

```python
import pandas as pd
```

```python
olympics = pd.read_csv('olympics.csv', skiprows=4)
```

__Using set_index()__

```python
# Set Athlete as index
olympics.set_index('Athlete', inplace=True)
```

```python
olympics.reset_index(inplace=True)
```

```python
ath = olympics.set_index('Athlete')
```

```python
ath.head()
```

__Using reset_index()__

```python
ath.reset_index(inplace=True)
```

```python
ath.head()
```

__Using sort_index()__

```python
ath.set_index('Athlete', inplace=True)
```

```python
ath.sort_index(inplace=True, ascending=False)
```

```python
ath.head()
```

__Using `loc[]`__

```python
ath.loc['BOLT, Usain']
```

```python
olympics[olympics.Athlete == 'BOLT, Usain']
```

```python
olympics.loc[olympics.Athlete == 'BOLT, Usain']
```

```python
olympics.loc[(olympics.NOC == 'USA') & (olympics.Medal == 'Gold')]
```

__Using `iloc[]`__

```python
olympics.iloc[1500]
```

```python
olympics.iloc[[1209,3902,2932]]
```

```python
olympics.iloc[1:4]
```

# Challenge

__Plot the total number of medals awarded at each of the Olympic games throughout history.__

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
len(olympics.Edition.value_counts())
```

```python
olympics.Edition.value_counts().sort_index().plot(kind='bar');
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Edition', data=olympics);
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='Edition', hue='Medal', data=olympics);
```
